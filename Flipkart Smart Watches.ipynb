# Scrapping Flipkart Website to find the price of Smart Watches

````python
import pandas as pd
from bs4 import BeautifulSoup
import requests
````

````python
url = "https://www.flipkart.com/search?q=smartwatches&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
source = requests.get(url)
soup1 = BeautifulSoup(source.text, 'lxml')
print(soup1.prettify())
````

## Scrapping the title of Smart Watches

````python
title = soup1.title.text
title
````

```python
p = soup1.select("div._4rR01T")
p
```

## Scrapping Names of all watches

````python
Watches = []
for watches in soup1.find_all('div', class_="_4rR01T"):
    Watches.append(watches.text)

Watches
````

## Scrapping MRP

````python
MRP = []
for price in soup1.find_all('div', class_='_3I9_wc _27UcVY'):
    MRP.append(price.text)

MRP
````

## Scrapping Offer Price

````python
Offer_Price = []
for price in soup1.find_all('div', class_='_30jeq3 _1_WHN1'):
    Offer_Price.append(price.text)

Offer_Price
````

## Getting the ratings

````python
Rating = []
for rating1 in soup1.find_all("div", class_="col col-7-12"):
    for rating in rating1.find_all('div', class_='_3LWZlK'):
        Rating.append(rating.text)
    
Rating
````

## Scrapping URL to buy Smart Watches

````python
URl = []
for url in soup1.find_all("a", class_="_1fQZEK"):
    URl.append("https://flipkart.com"+url['href'])
    
URl
````

## Finding Discount provided

````python
Discount = []
for offer in soup1.find_all('div', class_='col col-5-12 nlI3QM'):
    for off in offer.find_all("div", class_='_3Ay6Sb'):
        Discount.append(off.text)
    
Discount
````

## Exporting the data to csv file using Pandas Dataframe

````python
import pandas as pd
df = pd.DataFrame({'Watches': Watches, 'MRP': MRP, 'Offer_Price': Offer_Price,'Discount': Discount, 'URl': URl,})
df
````

````python
df.to_csv("Smart Watches Flipkart 1st Page.csv")
````

# Scrapping Multiple Pages

````python
pages = list(range(1, 25))
Watch = []
MRP = []
for page in pages:
    r = requests.get("https://www.flipkart.com/search?q=smartwatches&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}".format(page)).text
    soup2 = BeautifulSoup(r , 'lxml')
    title = soup2.find_all("div", class_="_4rR01T")
    for i in range (len(title)):
        Watch.append(title[i].text)
    mrp = soup2.find_all('div', class_="_30jeq3 _1_WHN1")
    for i in range(len(mrp)):
        MRP.append(mrp[i].text)
````

## Saving Daa using CSV Writer

````python
import csv
header = ['Watch', 'MRP']
data = [Watch, MRP]
with open("Smart Watches Flipkart Mul.csv", 'w', newline = '', encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
````
